```python
import numpy as np
import pandas as pd
import matplotlib as mlt
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
# Load image using PIL
img = Image.open("/content/pexels-pixabay-532310.jpg")
plt.imshow(img)
plt.show()
```

```python
img_np = np.array(img)
```

```python
img_np.shape
```

```python
pixels = img_np.reshape(-1 , 3)
```

```python
pixels.shape
```

```python
kmeans = KMeans(n_clusters=7 , init='k-means++' , n_init=12 , random_state=42)
```

```python
kmeans.fit(pixels)
```

```python
new_colors = kmeans.cluster_centers_[kmeans.labels_].astype('uint8')
```

```python
new_colors.shape
```

```python
compressed_img = new_colors.reshape(img_np.shape)
```

```python
plt.imshow(compressed_img)
plt.axis('off')
plt.show()
```
